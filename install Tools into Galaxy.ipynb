# 2 Become an Adimin
**To control Galaxy through the UI(installing tools, managing users, creating groups, etc), user must become an administrator.**  
Only registered users can become admins.  
To give a user admin privileges add the user's Galaxy login email to the configuration file `config/galaxy.yml`. 

## 2.1 Input codes in your terminal window:
```
vim config/galaxy.yml
```

## 2.2 Add user account at `admin_users:` and save this file, example as this picture:

<img src="imgs/admin_config.png" width = 80% height = 80% align=center />

## 2.3 Restart galaxy and log in the adminstrctor account, you will found there added `Admin` in the menu bar.

<img src="imgs/admin.png" width = 80% height = 80% align=center />


# 3 Connect Galaxy to a Tool Shed
## 3.1 Galaxy is by default connected to the Main Tool Shed. Here we build `tool_sheds_conf.xml` in order to add as many Tool Sheds: 

```
cp config/tool_sheds_conf.xml.sample config/tool_sheds_conf.xml
```

## 3.2 Edit `tool_sheds_conf.xml`:

```
vim config/tool_sheds_conf.xml
```

## 3.3 Add Test Tool Shed in `tool_sheds_conf.xml`, make this file as following codes:

```
<?xml version="1.0"?>
<tool_sheds>
    <tool_shed name="Galaxy Main Tool Shed" url="https://toolshed.g2.bx.psu.edu/"/>
<!-- Test Tool Shed should be used only for testing purposes.-->
    <tool_shed name="Galaxy Test Tool Shed" url="https://testtoolshed.g2.bx.psu.edu/"/>
</tool_sheds>
```


# 4 Install Mercurial
**To install tools from a Tool Shed you need to have [Mercurial](https://www.mercurial-scm.org/) available on the system that runs Galaxy.**  

Mercurial is a free, distributed source control management tool. It efficiently handles projects of any size and offers an easy and intuitive interface.

## 4.1 Check centos system version:

```
rpm -q centos-release
```

<img src="imgs/centos_version.png" width = 80% height = 80% align=center />

## 4.2 Download Mercurial

As you can see, the Lab server system is centos 7, so we download [Mercurial 5.4.2 for centos 7](https://www.mercurial-scm.org/release/centos7/mercurial-5.4.2-1.x86_64.rpm) with codes:

```
curl -O https://www.mercurial-scm.org/release/centos7/mercurial-5.4.2-1.x86_64.rpm
```

<img src="imgs/mercurial_1.png" width = 80% height = 80% align=center />

## 4.3 Install Mercurial

```
sudo rpm -ivh mercurial-5.4.2-1.x86_64.rpm
```

## 4.4 check installed Mercurial

```
rpm -qa|grep mercurial
```

<img src="imgs/mercurial_2.png" width = 80% height = 80% align=center />

**tips**:   
`rpm -qa|grep software_name` codes can help you check whether it installed this software, you just change `software_name` as the software you want to check.
